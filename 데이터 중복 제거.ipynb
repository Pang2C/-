```python
import pandas as pd
import os
import glob

# 1. 폴더 경로 설정
folder_path = "고가의 파일"  # 폴더 이름
output_file = "장윤서.csv"   # 저장할 파일 이름

# 2. 해당 폴더 내의 모든 .csv 파일 경로 가져오기
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# 3. 파일들을 순차적으로 읽어서 병합
df_list = []
for file in csv_files:
    try:
        df = pd.read_csv(file)
    except UnicodeDecodeError:
        df = pd.read_csv(file, encoding="ISO-8859-1")  # 인코딩 오류 대비
    df_list.append(df)

# 4. 하나로 병합
merged_df = pd.concat(df_list, ignore_index=True)

# 5. 완전 중복 제거 (모든 값이 동일한 행 제거)
deduped_df = merged_df.drop_duplicates()

# 6. 저장
deduped_df.to_csv(output_file, index=False)

print(f"{len(merged_df)} → {len(deduped_df)} (완전 중복 {len(merged_df) - len(deduped_df)}건 제거)")
```
